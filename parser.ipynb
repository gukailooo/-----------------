```python
def fibonacci_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

def generate_fibonacci_sequence(n):
    array = []
    for i in range(n + 1):
        array.append(fibonacci_recursive(i))
    return array

def generate_passwords_with_fibonacci(max_length=64):
    # Генерация первой 20 последовательности Фибоначчи, этого достаточно для генерации паролей
    fibonacci_sequence = generate_fibonacci_sequence(38)
    
    # Преобразуем числа Фибоначчи в строку
    fib_string = ''.join(str(num) for num in fibonacci_sequence)
    
    # Генерация паролей, каждый длиной до max_length символов
    passwords = []
    while len(fib_string) > 0:
        password = fib_string[:max_length]
        passwords.append(password)
        fib_string = fib_string[max_length:]
    
    return passwords

# Пример использования
passwords = generate_passwords_with_fibonacci()

# Выводим пароли
print("Generated Passwords:")
for password in passwords:
    print(password)
```

```python
def fibonacci_iterative(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    fib_sequence = [0, 1]
    for i in range(2, n + 1):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence

def generate_passwords_with_fibonacci(max_length=64):
    # Генерация последовательности Фибоначчи до 38-го элемента
    fibonacci_sequence = fibonacci_iterative(800)
    
    # Преобразуем числа Фибоначчи в строку
    fib_string = ''.join(str(num) for num in fibonacci_sequence)
    
    # Генерация паролей, каждый длиной до max_length символов
    passwords = [fib_string[i:i + max_length] for i in range(0, len(fib_string), max_length)]
    
    return passwords

# Пример использования
passwords = generate_passwords_with_fibonacci()

# Выводим пароли
print("Generated Passwords:")
for idx, password in enumerate(passwords, start=1):
    print(f'Password {idx}: {password}')
```

```python
def catalan_number(n):
    if n <= 1:
        return 1
    catalan = [0] * (n + 1)
    catalan[0] = catalan[1] = 1

    for i in range(2, n + 1):
        catalan[i] = sum(catalan[j] * catalan[i - j - 1] for j in range(i))
    return catalan[n]

def generate_passwords_with_catalan_numbers(max_length=64):
    # Генерация последовательности чисел Каталана
    catalan_sequence = [catalan_number(i) for i in range(1, 800)]
    
    # Преобразуем числа Каталана в строку и убираем точки
    catalan_string = ''.join(str(num) for num in catalan_sequence).replace('.', '')
    
    # Генерация паролей, каждый длиной до max_length символов
    passwords = [catalan_string[i:i + max_length] for i in range(0, len(catalan_string), max_length)]
    
    return passwords

# Пример использования
passwords = generate_passwords_with_catalan_numbers()

# Выводим пароли с счетчиком
print("Generated Passwords:")
for idx, password in enumerate(passwords, start=1):
    print(f"Password {idx}: {password}")
```

```python

import openpyxl

bradis_table = {}
for angle in range(0, 91):  # Углы от 0 до 90 градусов
    rad = math.radians(angle)
    bradis_table[angle] = {
        'sin': round(math.sin(rad), 8),
        'cos': round(math.cos(rad), 8),
        'tan': round(math.tan(rad), 8) if angle != 90 else 'inf',  # tan 90° -> бесконечность
        'cot': round(1 / math.tan(rad), 8) if angle != 0 else 'inf'  # cot 0° -> бесконечность
    }

# Создание новой книги и активного листа
wb = openpyxl.Workbook()
ws = wb.active

# Назначение заголовков столбцов
ws.append(["Угол (градусы)", "sin", "cos", "tan", "cot"])

# Заполнение данными из словаря bradis_table
for angle, values in bradis_table.items():
    ws.append([angle, values['sin'], values['cos'], values['tan'], values['cot']])

# Сохранение файла
file_path = "bradis_table.xlsx"
wb.save(file_path)
```

```python
from PyQt5.QtWidgets import QApplication, QMainWindow, QSpinBox, QPushButton, QVBoxLayout, QWidget, QMessageBox

class PasswordGenerator:
    def generate_sequence(self, start, ratio, n):
        return [start * ratio ** i for i in range(n)]

    def generate_passwords_with_geometric_progression(self, start, ratio, n_terms, number_passwords, max_length=64):
        try:
            if number_passwords < 1:
                raise ValueError("Количество паролей должно быть не менее 1.")

            sequence = self.generate_sequence(start=start, ratio=ratio, n=n_terms)
            sequence_string = ''.join(str(num) for num in sequence)
            passwords = [sequence_string[i:i + max_length] for i in range(0, len(sequence_string), max_length)]
            passwords = passwords[:number_passwords]
                    
            with open('geometric_progression_passwords.txt', 'w') as file:
                file.write('Каждый новый пароль - это 64 символа геометрической прогрессии, '
                           'причем начало {}, знаменатель прогрессии {}, количество членов прогрессии {}:\n'.format(start, ratio, n_terms))
                for idx, password in enumerate(passwords, start=1):
                    file.write(f"Password with geometric progression {idx}: {password}\n")
            
            QMessageBox.information(None, 'Уведомление', 'Пароли успешно записаны.')
        
        except Exception as e:
            QMessageBox.critical(None, 'Ошибка', f'Произошла ошибка: {e}')

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Geometric Progression Password Generator')
        
        self.spin_box = QSpinBox(self)
        self.spin_box.setRange(1, 100)
        self.spin_box.setValue(5)
        
        self.generate_button = QPushButton('Generate Passwords', self)
        self.generate_button.clicked.connect(self.generate_passwords)
        
        layout = QVBoxLayout()
        layout.addWidget(self.spin_box)
        layout.addWidget(self.generate_button)
        
        container = QWidget()
        container.setLayout(layout)
        self.setCentralWidget(container)
        
        self.password_generator = PasswordGenerator()

    def generate_passwords(self):
        number_passwords = self.spin_box.value()
        start = 2
        ratio = 3
        n_terms = 500
        self.password_generator.generate_passwords_with_geometric_progression(start, ratio, n_terms, number_passwords)

if __name__ == '__main__':
    import sys
    app = QApplication(sys.argv)
    mainWindow = MainWindow()
    mainWindow.show()
    sys.exit(app.exec_())
```
